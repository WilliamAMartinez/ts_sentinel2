# Animation NDVI

The attempt of this script is to create an animation of the ndvi images for the year 2017, so that we can evaluate possible changes bewtween different classes. As an example I bring the class water with subcategory of land use Estuaries. This category forms a transition zone between river environments and maritime environments. Depending on the river flow this areas can have diverse land cover typer over the year, usually from water to non-vegetated.

```python
from osgeo import gdal, ogr
import numpy as np
import os
from matplotlib import pyplot as plt
import ipywidgets as wg
```

```python
ndvi_directory = r'/home/user/Documents/TESISMASTER/IMAGES/TO_PROCESS_10m/NDVI_10m_clip'
#importing the paths of the raster files
list_path_images = [os.path.join(ndvi_directory,f) for f in os.listdir(ndvi_directory) if f.endswith('.tif')]
list_path_images.sort()
#dates of the imagery
day_shoot = ["2017/01/15", "2017/04/05", "2017/05/25", "2017/06/04", "2017/06/14",
              "2017/07/04", "2017/07/09", "2017/07/14", "2017/07/24", "2017/07/29",
              "2017/08/03", "2017/08/08", "2017/08/13", "2017/08/18", "2017/09/02" ,
              "2017/09/07", "2017/09/12", "2017/09/22", "2017/09/27", "2017/10/02",
              "2017/10/12", "2017/10/22", "2017/10/27", "2017/11/11", "2017/11/16",
              "2017/11/21", "2017/12/01","2017/12/16", "2017/12/21"]
```

In the following block I will import the ndvi images one by one using a loop

```python
bands_data= []
for i in list_path_images:
    data_source = gdal.Open(i)
    r_band = data_source.GetRasterBand(1)
    band = r_band.ReadAsArray()
    #appending in a list the raster
    bands_data.append(band)
    #Projection and geotransformation are part of the last layer, so I will overwrite the object
    gt = data_source.GetGeoTransform()
    proj = data_source.GetProjectionRef()
    xs =  data_source.RasterXSize   # Number of columns
    ys =  data_source.RasterYSize   # NUmber of rows
#stacking layers
bands_data = np.dstack(bands_data)
data_source = None
#printing dimensions of the array
print(bands_data.shape)
```

So I will plot the rasters of NDVI with the points associated with water

```python
shapefile = r'/home/user/Documents/TESISMASTER/VECTOR/CartoBase/Points_water_ndvi.shp'
driver =  ogr.GetDriverByName("ESRI Shapefile")
data_source = driver.Open(shapefile)
layer = data_source.GetLayer()
x = []
y = []
for feat in layer:
    field = feat.GetField(1)
    if field == 'Water':
        pt = feat.geometry()
        x.append(pt.GetX())
        y.append(pt.GetY())
```

```python
#before showing the raster I want to define the grid of the map
minx =  gt[0]
maxy =  gt[3]
maxx = minx + gt[1]*xs
miny = maxy + gt[5]*ys
# slider
c_slide = wg.IntSlider(value = 0, min= 0, max = 28, step =1)
def plot_raster(a):
    f, axarr = plt.subplots(1, figsize=(7,7))
    #use imshow to plot the raster
    et = (minx, maxx, miny, maxy) #passing extent to imshow
    im = axarr.imshow(bands_data[:,:,a], vmin=-1, vmax=1, cmap = "hot", extent= et)
    scat = axarr.scatter(x,y,s=1)
    #add the legend
    cb = plt.colorbar(im,spacing= 'proportional')
    cb.set_label("NDVI")
    print(f"Date: {day_shoot[a]}")
#interaction slider
wg.interact(plot_raster, a = c_slide)
```
